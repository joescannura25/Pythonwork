```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
df3 = pd.read_csv('new_df.csv',  index_col = 'Unnamed: 0')
```

```python
df3
```

```python
df4 = df3.groupby('state').sum()
```

Here we group the different statistics by state to compute state totals for tweets, retweets, favorites, and followers. 

```python
state_list = df4['tweets_count']
follower_count = df4['follower_count']
retweet_count = df4['retweet_count']
favorite_count = df4['favorite_count']
```

Here we create a cimple list of only the state total counts for each category. 

```python
state_tweet_count = state_list.values
state_follower_count = follower_count.values
state_retweet_count = retweet_count.values
state_favorite_count = favorite_count.values
```

Then we pull only the values. 

```python
i = 0
state_tweets1 = []
while i < len(state_tweet_count):
    state_tweets1.append(state_tweet_count[i])
    state_tweets1.append(state_tweet_count[i])
    i += 1
    
i = 0
state_followers1 = []
while i < len(state_follower_count):
    state_followers1.append(state_follower_count[i])
    state_followers1.append(state_follower_count[i])
    i += 1
    
i = 0
state_retweets1 = []
while i < len(state_retweet_count):
    state_retweets1.append(state_retweet_count[i])
    state_retweets1.append(state_retweet_count[i])
    i += 1
    
i = 0
state_favorites1 = []
while i < len(state_favorite_count):
    state_favorites1.append(state_favorite_count[i])
    state_favorites1.append(state_favorite_count[i])
    i += 1
```

Then we duplicate them so they can be put be combined with the original dataframe for each candidate. 

```python
df5= df3.sort('state')
df5.head()
```

```python
df5['statetotal_tweets'] = state_tweets1
```

```python
df5['statetotal_followers'] = state_followers1
df5['statetotal_retweets'] = state_retweets1
df5['statetotal_favorites'] = state_favorites1
```

Creates the new column in the original dataframe for state totals. 

```python
df5['%of_tweets'] = df5['tweets_count']/df5['statetotal_tweets']
```

```python
df5['%of_followers'] = df5['follower_count'] / df5['statetotal_followers']
df5['%of_retweets'] = df5['retweet_count'] / df5 ['statetotal_retweets']
df5['%of_favorites'] = df5['favorite_count'] / df5 ['statetotal_favorites']
```

Computes % of overall state # for each category for each candidate.
