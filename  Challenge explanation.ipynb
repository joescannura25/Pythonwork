## Challenge Area

Our challenege area was in the **Study of Social Networks**. We decided to focus on studying **Twitter** data and using Twitter's Rest API's to study the timelines of political candidates in the U.S. 

We used **Matthew A. Russell's** book **"Mining the Social Web"** as a guide to using and studying twitter data. The book has a very good basic walkthrough on setting up a developer account with twitter and receiveing the necessary authorizations to retrieve data using your developer account. 

Gathering information through the API requires *4 different authentification codes, including OAuth codes*, which allow you to make a **limited number** of API calls through your account to retrieve various sorts of information from twitter. 

The next challenging area of using the API data is to breakdown the actual data stream which it provides for the content you require.
